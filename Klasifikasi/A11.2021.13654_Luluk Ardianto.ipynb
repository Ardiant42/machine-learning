6. Isi _code-cell_ dibawah dengan kode untuk:
    - **Ubah label _species, island dan sex_** menjadi angka
    - **Masukkan hasil labelling** kedalam dataset


```python
# Buatlah dictionary dengan isi sebagai berikut:

# kamusSpesies = {str(TARGET) : INDEX for INDEX, TARGET in enumerate('''FIX ME''')}
# kamusIsland = {str(TARGET) : INDEX for INDEX, TARGET in enumerate('''FIX ME''')}
# kamusSex = {str(TARGET) : INDEX for INDEX, TARGET in enumerate('''FIX ME''')}

# Contoh kolom dalam DataFrame
spesies = df['species'].unique()  # Daftar unik dari kolom 'species'
island = df['island'].unique()   # Daftar unik dari kolom 'island'
sex = df['sex'].unique()         # Daftar unik dari kolom 'sex'

# Membuat dictionary dengan indeks dari enumerate
kamusSpesies = {str(target): index for index, target in enumerate(spesies)}
kamusIsland = {str(target): index for index, target in enumerate(island)}
kamusSex = {str(target): index for index, target in enumerate(sex)}

print("Dictionary Spesies:", kamusSpesies)
print("Dictionary Island:", kamusIsland)
print("Dictionary Sex:", kamusSex)
```

output yang diharapkan :      



```python
# Mapping dictionary yang telah dibuat ke dalam data frame dan lihat hasil
# '''FIX ME''' = '''FIX ME'''.map(kamusSpesies)
# '''FIX ME''' = '''FIX ME'''.map(kamusIsland)
# '''FIX ME''' = '''FIX ME'''.map(kamusSex)

# Mapping dictionary ke dalam DataFrame
# Mapping dictionary ke dalam kolom DataFrame
df['species'] = df['species'].map(kamusSpesies)  # Mapping untuk 'species'
df['island'] = df['island'].map(kamusIsland)  # Mapping untuk 'island'
df['sex'] = df['sex'].map(kamusSex)  # Mapping untuk 'sex'

# Menampilkan hasil setelah mapping
print("DataFrame dengan hasil mapping:")
print(df.head())
```

## **FUNCTION YANG BOLEH DIGUNAKAN JIKA MAU**

```python
# Function
def trainSemua(x_train, y_train):
    knn = KNeighborsClassifier()
    dt = DecisionTreeClassifier()
    nb = GaussianNB()
    knn.fit(x_train, y_train)
    dt.fit(x_train, y_train)
    nb.fit(x_train, y_train)
    return knn, dt, nb

def predikSemua(knn, dt, nb, x_test, y_test):
    knnPred = knn.predict(x_test)
    dtPred = dt.predict(x_test)
    nbPred = nb.predict(x_test)
    knnAcc = accuracy_score(y_test, knnPred)
    dtAcc = accuracy_score(y_test, dtPred)
    nbAcc = accuracy_score(y_test, nbPred)
    return knnAcc, dtAcc, nbAcc
```

## **MODELLING**

8. **Bagi dataset** menjadi _train-set_ dan _test-set_ dengan ketentuan sebagai berikut:

    a. X = seluruh kolom fitur kecuali kolom target

    b. y = kolom target
    
    c. Bagi dengan rasio 70% _train-set_ dan 30% _test-set_


```python
# Bagi menjadi fitur dan target
from sklearn.model_selection import train_test_split

# Mengasumsikan bahwa kolom 'target' adalah kolom target
# Jika nama kolom target berbeda, ganti dengan nama kolom yang sesuai
target_column = 'sex'  

# Memisahkan fitur (X) dan target (y)
X = df.drop(columns=[target_column])  # Seluruh kolom kecuali 'target'
y = df[target_column]  # Kolom target
```

```python
# Split menjadi train dan test sesuai rasio yang disebutkan
# Membagi dataset menjadi train-set dan test-set dengan rasio 70/30
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print("Ukuran train-set:", x_train.shape, y_train.shape)
print("Ukuran test-set:", x_test.shape, y_test.shape)
```

9. Buatlah perbandingan model menggunakan algoritma **klasifikasi** antara 2 data yaitu

    a. Data yang di  **NORMALISASI**

    b. Data yang  **BELUM DI NORMALISASI**

    kemudian tampilkan hasil akurasi dari dua dataset tersebut

```python
# Cell berikut untuk data yang belum diNormalisasi
# Boleh mengikuti arahan berikut atau menggunakan cara sendiri
# Impor library yang diperlukan
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
```

```python
# Melatih model pada data yang belum dinormalisasi
knn, dt, nb = trainSemua(x_train, y_train)
```

```python
# Mendapatkan akurasi pada data yang belum dinormalisasi
knnAcc, dtAcc, nbAcc = predikSemua(knn, dt, nb, x_test, y_test)
```

```python
# Deklarasi Scaler dan scale fitur
# Normalisasi data menggunakan StandardScaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()  # Bisa juga menggunakan MinMaxScaler atau lainnya
X_norm = scaler.fit_transform(X)
```

```python
# Split ulang menjadi train dan test sesuai rasio yang disebutkan.
# Kali ini menggunakan fitur X yang sudah dinormalisasi
# Split ulang menjadi train-set dan test-set dengan data yang sudah dinormalisasi
x_trainN, x_testN, y_trainN, y_testN = train_test_split(X_norm, y, test_size=0.3, random_state=42)
```

```python
# Cell berikut untuk data yang sudah diNormalisasi
# Boleh mengikuti arahan berikut atau menggunakan cara sendiri
```

```python
# Melatih model pada data yang sudah dinormalisasi
knnN, dtN, nbN = trainSemua(x_trainN, y_trainN)
```

```python
# Mendapatkan akurasi pada data yang sudah dinormalisasi
knnAccN, dtAccN, nbAccN = predikSemua(knnN, dtN, nbN, x_testN, y_testN)

# Menampilkan hasil akurasi
print("Akurasi model pada data yang sudah dinormalisasi:")
print("KNN:", knnAccN, "Decision Tree:", dtAccN, "Naive Bayes:", nbAccN)
```
